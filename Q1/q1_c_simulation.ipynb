# Creating a Simulated Sample of 100,000 stars for Euclid Mission 

This notebook shows the method of simulating a sample of 100,000 stars for the Euclid programme. This notebook assumes that a sample of stars has been given with their 'Y-J' and 'J-H' colour indices already calculated. It also assumes that the provided sample of stars is a representative sample of the real distribution of such stars. Hence, we first need to recreate that distribution by Kernel Density Estimation and then a random sample of 100,000 stars can be extracted from this distribution. 

First we read the csv file containing the sample.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KernelDensity
sns.set()
%matplotlib inline
```

```python
t = pd.read_csv('YJHcatalog.csv')
```

Let us now visualize the data.

```python
x = t['YJ'].values #.values is to convert the pd dataframe to a numpy array with values
y = t['JH'].values
plt.scatter(x,y)
plt.xlabel('Y-J')
plt.ylabel('J-H')
plt.show()
```

We now proceed to do a Kernel Density estimation of the data. The code used here is the code used as part of the Kernel Density Estimation exercise as part of Assignment 3. I use the gaussian kernel here with a sufficiently small bandwidth to preserve detail. Doing a gridsearch using cross validation on such a large dataset is going to take a lot of time (I actually attempted it.), hence by trial and error I assume a bandwidth of 0.5.

```python
xy = np.vstack([x,y])
xmin = x.min()
xmax = x.max()
ymin = y.min()
ymax = y.max()
kde = KernelDensity(bandwidth = 0.5, kernel = 'gaussian') # Only Gaussian and Tophat kernels are available for random sampling
kde.fit(xy.T)
Nx = 500
Ny = 1000
xgrid = np.linspace(xmin, xmax, Nx)
ygrid = np.linspace(ymin, ymax, Ny)
mesh = np.meshgrid(xgrid, ygrid)
```

```python
tmp = map(np.ravel, mesh)
Xgrid = np.vstack(tmp).T
log_dens = kde.score_samples(Xgrid)
dens1 = x.shape*np.exp(log_dens).reshape((Ny, Nx))
```

```python
plt.imshow(dens1.T, cmap=plt.cm.hot) # Used cm.hot to provide the necessary contrast between distribution and background
```

Having obtained the probability density function (which looks like a 2D Gaussian) by fitting the values, we now proceed to simulate 100000 values as is our requirement.

```python
s = kde.sample(100000)
s # Visulaize the simulated data
```

```python
s.shape #As is expected
```

```python
xs, ys = np.hsplit(s,2) #For plotting purposes, it is required to split the array into two columns
```

```python
xs # Visualize the Y-J data
```

```python
ys
```

```python
plt.scatter(xs,ys)
plt.xlabel('Simulated values of Y-J')
plt.ylabel('Simulated values of J-H')
plt.show()
```

It is easily seen that these values approximately resemble our original sample in distribution although there are a lot more stars. Thus, this sample can be taken as a representative sample for the Euclid mission and science analysis can be done using this.
